# Casos confirmados de COVID-19

```python
import pandas as pd
from urllib.request import urlretrieve # Biblioteca utilizada para buscar a URL dos dados
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Importando  os dados do github - https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv
# Devemos abrir o arquivo e copiar a URL
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
# Criando o arquivo .csv com os dados
urlretrieve(url, 'global_cases_covid19.csv')
```

```python
# Carregando os dados
df_covid = pd.read_csv('global_cases_covid19.csv')
```

```python
# Deletando as colunas Lat e Long, pois não são necessárias
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True) # o axis indica que é uma coluna
```

```python
# Agrupamento dos paises
df_country = df_covid.groupby('Country/Region').sum()
```

```python
# Visualizando o Brasil
df_country.loc['Brazil']
```

```python
### pegando os indices
date = df_country.loc['Brazil'].index
### pegando os valores
cases = df_country.loc['Brazil'].values
```

```python
plt.figure(figsize=(14,8)) # define o tamanho do gráfico
plt.bar(date, cases)
```

```python
s_brasil = df_country.loc['Brazil']
```

```python
s_brasil = s_brasil[s_brasil > 0]
```

```python
tam = len(s_brasil)

plt.figure(figsize=[14,8])
plt.xticks(rotation=60)
plt.bar(s_brasil.index[tam-100:tam], s_brasil.values[tam-100:tam])
plt.title('Total de casos confirmados de COVID-19 no Brasil')
```
